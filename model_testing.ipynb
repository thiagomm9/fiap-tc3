```python
# Define base path

BASE_PATH = '/content/drive/MyDrive/AI/tcfase3'
```

```python
# Nacessary Imports

from transformers import T5Tokenizer, T5ForConditionalGeneration
```

```python
# Load base model from the disk

MODEL_PATH = BASE_PATH + '/base_model'
base_model = T5ForConditionalGeneration.from_pretrained(MODEL_PATH)
base_tokenizer = T5Tokenizer.from_pretrained(MODEL_PATH)
```

```python
# Load fine tuned model from the disk

MODEL_PATH = BASE_PATH + '/fine_tuned_model'
trained_model = T5ForConditionalGeneration.from_pretrained(MODEL_PATH)
trained_tokenizer = T5Tokenizer.from_pretrained(MODEL_PATH)
```

```python
# Define the generate answer function

def generate_answer(product):
  input_ids = base_tokenizer(f"Please describe this product: {product}", return_tensors='pt').input_ids.to("cuda")
  outputs = base_model.to("cuda").generate(input_ids)
  base_answer = base_tokenizer.decode(outputs[0], skip_special_tokens=True)
  input_ids = trained_tokenizer(f"Please describe this product: {product}", return_tensors='pt').input_ids.to("cuda")
  outputs = trained_model.to("cuda").generate(input_ids)
  trained_answer = trained_tokenizer.decode(outputs[0], skip_special_tokens=True)
  print(f'Please describe this product: {product}')
  print(f'Base model answer: {base_answer}')
  print(f'Trained model answer: {trained_answer} \n ------------------------- \n')
```

```python
product = "A Day in the Life of China"
generate_answer(product)
product = "On Happiness, U.S. Edition"
generate_answer(product)
product = "The Official CIA Manual of Trickery and Deception"
generate_answer(product)
product = "The Devils of Loudun"
generate_answer(product)
product = "The Big Book of Stress Relief Games: Quick, Fun Activities for Feeling Better"
generate_answer(product)
product = "Coloring for Grown-Ups Holiday Fun Book"
generate_answer(product)
product = "Brothers of the Knight (Picture Puffin Books)"
generate_answer(product)
product = "Dearest Love (Best of Betty Neels)"
generate_answer(product)
product = "The Dream Unfolds (Crosslyn Rise, Book 2)"
generate_answer(product)
product = "eBay PowerSeller Business Practices For Dummies"
generate_answer(product)
product = "Essentials of Banking"
generate_answer(product)
product = "TKO Hiring!: Ten Knockout Strategies for Recruiting, Interviewing, and Hiring Great People"
generate_answer(product)
product = "The Bleeding Land"
generate_answer(product)
```
